```python
# Carregando libs
import statsmodels.api   as sm
import pandas            as pd
import matplotlib.pyplot as plt
import numpy             as np
import seaborn           as sb
from   statsmodels.graphics.tsaplots      import plot_acf
from   statsmodels.graphics.tsaplots      import plot_pacf
from   statsmodels.tsa.statespace.sarimax import SARIMAX
from scipy import stats 
import statistics 

# Carregando e checando estrutura de dados
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'], index_col='date')
df.head()
df.describe()
df.info()

# Definindo a funcao de analise exploratoria dos dados
def analiseExploratoria(df,x,y,lags):
    fig, axes = plt.subplots(nrows=2, ncols =3, figsize=(12, 12))
    fig.tight_layout()
    axes[0,0].plot(df)
    axes[0,1].boxplot(df)
    sb.histplot(y, ax = axes[0,2], kde=True)
    plot_acf(x=x, lags=lags, ax= axes[1,0])
    plot_pacf(x=x,lags=lags, ax= axes[1,1])    
    
analiseExploratoria(df=df,x= df.index,y=df.value,lags=60)    

df_diff = df.diff(periods = 1).dropna()

analiseExploratoria(df=df_diff,x= df_diff.index,y=df_diff.value,lags=60)  

# Ajustando o modelo e agrupando as metricas de performance
df_treino = df_diff.iloc[0:200,0]
df_teste  = df_diff.iloc[200:204,0]

df_teste
```

```python
# Loop que testa varias combinacoes de ordem do modelo e guarda as metricas num dataframe
metricas = pd.DataFrame(np.zeros((5*5,5), dtype=float))
contador = 0
metricas.rename(columns = {'0':'TEST', '1':'ODI', '2':'T20'}, inplace = True) 
for p in range(5):
    for q in range(5):
    # Colocamos esse "se" para pular um modelo nulo
        if p == 0 and q ==0:
                continue
        contador += 1
        modelo = SARIMAX(endog = df_treino, order=(p,0,q), seasonal_order = (0,0,0,0)).fit()
        previsoes = modelo.predict('2008-04-01', '2008-06-01')
        EQMP = statistics.mean((df_teste - previsoes)**2)
        metricas.iloc[contador,0] = p
        metricas.iloc[contador,1] = q
        metricas.iloc[contador,2] = modelo.aic
        metricas.iloc[contador,3] = modelo.bic
        metricas.iloc[contador,4] = EQMP

        
# Agora podemos comparar os 10 modelos
metricas

# Podemos notar que um ARMA(1,1) teve o mínimo erro de previsão
modelo = SARIMAX(endog = df_treino, order=(2,0,0), seasonal_order = (0,0,0,0)).fit()
modelo.plot_diagnostics(figsize = (12, 7), lags=20)
```

```python
modelo = SARIMAX(endog = df_treino, order=(1,0,1), seasonal_order = (0,0,0,0)).fit()
previsoes = modelo.predict('2008-04-01', '2008-06-01')
previsoes 
```
